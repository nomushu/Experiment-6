## 2 Experiment setups
### Accesories
- Thermometer 
- Thermostat
- Manometer
- Capillary with managable mercury levei inside

### Ecperiment procedure
Set the desired temperature with thermostat. Before we start varying the volume inside the cappillary, the temperature surrounding must reach to a stable value. The measurement of pressure inside the capillary must be given for decided volume interval. In our experiment, the volume was managed by the handwheel placed under the capillary. Note that mercury is toxic substance so that we have to take care of leak out of it by keeping the pressure in the capillary under 50 Pa through the experiment. We start measurement from room temperature and repeat the process for $33C\degree$, $36C\degree$, $39C\degree$, $42C\degree$, $45C\degree$, $47C\degree$ and $50C\degree$.
