## 3. Task 2 Isotherms and Maxwell lines
### Background theorem
Assume that sulfur hexaflouride as a real gas that follows Real gas equation
$$\left(p+a{(\frac{n}{V})^2}\right)\left(\frac{V}{n}-b\right) =RT \tag{1}$$
- $p$ = pressure of the gas (Pa)
- $V$ = volume of the gas (L)
- $R$ = ideal gas constant 8.31
- $T$ = temperature of the gas (K)
- $n$ = amount of substance (mol)
- $a$,$b$ = fixed value 

Pressure as a subject,
$$p_{(V)} = \frac{RT}{(\frac{V}{n}-b)} - a\left(\frac{n}{V}\right)^2 \tag{2}$$

This formula must give local extremas at the temperature below the critical temperature $T_k$.</br>
We also define $p_s$ as a saturation vapour pressure that is the height Maxwell line is placed.
The maximum volume for the extrema is defined as $V_{g}$ and for minimum extrema is defined as $V_{l}$.
From the definition of the Maxwell line,
$$\int_{\Delta V} (p_{(V)} - p_s) \:dV = \int_{V_g}^{V_l} \left[ \frac{RT}{(\frac{V}{n}-b)} - a(\frac{n}{V})^2 - p_s \right]_{T>T_k}\:dV = 0 \tag{3}$$

There are no Maxwell lines above the critical Temperature.</br>
The theoretical value of critical temperature is given by
$$T_k = \frac{8a}{27Rb} \tag{4}$$
using specific values for $a$ and $b$ of sulfur hexaflouride.

## 4. Task 3 Satureaion vapour pressure against temperatire
### Backgrount theorem
We know vapour-pressure equation that is a differential equation of vapour pressure.
$$\frac{dp_s}{dT} = \frac{p_sQ_{23}}{RT^2} \tag{5}$$
$Q_{23}$ is molar latent heat of evaporation.</br>
This differential equation is solved by to
$$p_s = p_{s0}\exp{\left[-\frac{Q_{23}}{RT}\right]} \tag{6}$$
and we do natural logarithm to both side 
$$\ln{(p_s)} = \ln{(p_{s0})} -\frac{Q_{23}}{RT} \tag{7}$$
so it becomes linear function with respect to $(\ln{p_s})$ and $\left(\frac{1}{T}\right)$ as $Q_{23}$ and $p_{s0}$ are staying constant.
